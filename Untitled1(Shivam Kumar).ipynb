```python
import numpy as np
import matplotlib.pyplot as plt
```

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+. ... SciPy makes use of Matplotlib.

```python
plt.rcParams['agg.path.chunksize'] = 1000
plt.title('2d plot of the data to show the variation b/w equity index and gold price(Mumbai)')
plt.xlabel('Gold_Mumbai')
plt.ylabel('BSE_Index')
plt.plot(project['Gold_Mumbai'], project['BSE_Index'], color='blue')
plt.show()
```

```python
plt.rcParams['agg.path.chunksize'] = 1000
plt.title('2d plot of the data to show the variation b/w equity index and oil price')
plt.xlabel('Cost_of_oil')
plt.ylabel('BSE_Index')
plt.plot(project['Cost_of_oil'], project['BSE_Index'], color='blue')
plt.show()
```

As we can see from the above two 2d curves that slope of the curve between equity index and Gold price (mumbai) is somewhat increasing gradually or is larger as compared to the slope of the curve between Equity index and Oil price.

So by increasing slope or bigger slope means the value of gold price has increased as the Equity index increases. But the oil price was not always increasing , after some moment the oil price decreased. So by this observation I can give advice to Ms Ania to invest in Gold price rather than investing in Oil price as by observing the historical data we can say that Gold price is increasing as Equity index increases.
